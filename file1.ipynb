```python
from tensorflow.keras import Sequential, layers

# Create the model
model = Sequential()

# Add layers
model.add(layers.Dense(40, input_shape=(7,), activation='softmax'))  # Input layer with softmax
model.add(layers.Dense(40, activation='relu'))
model.add(layers.Dense(30, activation='sigmoid'))
model.add(layers.Dense(25, activation='relu'))
model.add(layers.Dense(59, activation='sigmoid'))  # 22 neurons for 22 classes

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',  # Use this if y_train is integer encoded
              metrics=['accuracy'])

# Fit the model
model.fit(x_train, y_train, epochs=500)
```
